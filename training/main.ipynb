# Handwritten Character Recognition - Training

Convolutional Neural Network (CNN) for EMNIST character classification.

## 1. Setup

```python
import sys
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix

from config import (
    ARCHITECTURE, EPOCHS, BATCH_SIZE, VALIDATION_SPLIT,
    LOSS_FUNCTION, OPTIMIZER, LEARNING_RATE, METRICS, LABELS,
)
from libs import (
    load_and_preprocess,
    build_model,
    compile_model,
    train_model,
    evaluate_model,
    save_model_keras,
    print_evaluation_report,
    get_most_confused_pairs,
    get_label,
)
```

## 2. Load and Explore Data

```python
x_train, y_train, x_test, y_test = load_and_preprocess()

print(f"Training set: {x_train.shape}, {y_train.shape}")
print(f"Test set:     {x_test.shape}, {y_test.shape}")
print(f"Classes:      {y_train.shape[1]}")
```

```python
fig, axes = plt.subplots(2, 5, figsize=(12, 5))
for i, ax in enumerate(axes.flat):
    idx = np.random.randint(len(x_train))
    ax.imshow(x_train[idx].squeeze(), cmap='gray')
    ax.set_title(get_label(np.argmax(y_train[idx])))
    ax.axis('off')
plt.tight_layout()
plt.show()
```

## 3. Build Model

```python
model = build_model()
model = compile_model(model, learning_rate=LEARNING_RATE)
model.summary()
```

## 4. Train

```python
history = train_model(
    model,
    x_train,
    y_train,
    epochs=EPOCHS,
    batch_size=BATCH_SIZE,
    validation_split=VALIDATION_SPLIT,
)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

ax1.plot(history.history['accuracy'], label='Train')
ax1.plot(history.history['val_accuracy'], label='Validation')
ax1.set_title('Accuracy')
ax1.set_xlabel('Epoch')
ax1.legend()

ax2.plot(history.history['loss'], label='Train')
ax2.plot(history.history['val_loss'], label='Validation')
ax2.set_title('Loss')
ax2.set_xlabel('Epoch')
ax2.legend()

plt.tight_layout()
plt.show()
```

## 5. Evaluation

```python
results = evaluate_model(model, x_test, y_test)
print_evaluation_report(results)
```

```python
cm = results['confusion_matrix']

plt.figure(figsize=(14, 12))
sns.heatmap(cm, annot=False, fmt='d', cmap='Blues',
            xticklabels=LABELS, yticklabels=LABELS)
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.tight_layout()
plt.show()
```

## 6. Confusion Analysis

```python
confused = get_most_confused_pairs(cm, top_n=15)
print("Top 15 confused pairs (true -> predicted):")
print("-" * 40)
for true_label, pred_label, count in confused:
    print(f"  '{true_label}' -> '{pred_label}': {count}")
```
